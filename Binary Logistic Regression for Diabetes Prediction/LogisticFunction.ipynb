## The logistic Function

```python
import math as math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
def logistic(x):
    return 1.0 / (1+ math.exp(-x))
# Calcuate logisitic value
```

```python
logistic(1)
# Function taken from 'Data Science from Stratch' by Joel Grus
```

```python
def logistic_prime(x):
    return logistic(x) * (1 - logistic(x))
# Calculating logistic deriative 
```

```python
logistic_prime(1)
```

```python
fig = plt.figure(figsize = [16, 8]);
sns.lineplot(data=logistic, x='X', y='Y', color='black');
plt.title('The Logistic Function', size=25)
plt.xlabel('Logit(P)', size=20);
plt.xticks(size=15);
plt.yticks(size=15);
plt.text(-8.5, .8, r'$\sigma(x) = \frac{1}{1 + exp^{-x}}$', size = 25);
plt.axvline(0, 0,1, c='black');
plt.axhline(0.5, -10,10, c='black');
plt.ylabel('Probability', size=20);
plt.savefig('LogisticFunction.png', bbox_inches='tight')
```
